## 1. tuple(튜플)

### 튜플의 경우 리스트와 비슷하게 생겼지만(리스트: 대괄호, 튜플: 소괄호)
#### 리스트의 경우 .append() 등 함수를 사용하여 변수안의 요소를 추가(삭제) 할 수 있지만  튜플의 경우 ➡︎ &nbsp; *'변경이 불가'* 하다.

#### ➥ Tuple(튜플)은 *값을 고정할 수 있는 것* 이라고 생각하면 된다.(추가, 삭제, 변경 등이 안 된다.)

```python
from os import name

t1 = (1,2, 'a','b')
del t1[0]
```

#### TypeError: 'tuple' object doesn't support item deletion. 튜플은 삭제할 수 없다고 안내하고 있다.

```python
t1 = (1,2,'a','b')
t1[0] = 'c'
```

#### ➥ List 였다면 0번째, 즉 첫번째 1이 c로 바뀌었겠지만 Tuple은 그럴수 없다.But *'인덱싱, 슬라이싱'* 은 가능하다. 변경은 안 되지만 볼수는 있다.

```python
t1 = (1,2,'a','b')
print(t1[0:3])
```

#### ⇧ 🪚슬라이싱은 가능하다.(O)

```python
t1 = (1,2,'a','b')
print(t1[0])
```

#### ⇧ 🗂️인덱싱은 가능하다.(o)

#### ※ 튜플의 경우도 📟 연산(+−×÷)은 가능하다.(o)

```python
t1 = (1,2,'a','b')
t2 = (3,4)
print(t1+t2)
```

### 🔥1-1. 튜플(Tuple) 속 가변 객체의 비밀

* **핵심 명제:** "튜플을 생성하고 나면 튜플에 담긴 참조(주소 쪽지)는 변경할 수 없지만, 참조하고 있는 객체 자체는 변경할 수 있다."
* **실험과 검증:** `t = (1, [2, 3], 4)`
  1. `t[1] = [99]` ➞ **에러 (Failure)** : 주소 쪽지 자체를 갈아끼우는 대입 연산(`=`)이므로 금지.
  2. `t[1].append(5)` ➞ **성공 (Success)** : 주소를 타고 들어가 방 내부를 수정한 것이며, 튜플 안의 주소 쪽지는 변함없으므로 허용.

### 🔥1-2 "튜플 안의 객체는 변경할 수 있다.
##### 이는 무언가 튜플 '안에' 있다고 하는 것이 실제로 일어나는 일과 정확히 일치하지 않기 때문이다. 더 정확히 말하면 '튜플을 생성하고 나면 튜플에 담긴 참조는 변경할 수 없지만, 참조하고 있는 객체는 변경할 수 있다.'라고 해야 한다.
##### 왜 튜플이 있는 걸까. 이는 튜플이 어떤 연산은 더 효율적으로 만들어 주고, 또 어떤 연산은 안전하게 만들어 주기 때문이다." 라고 제니퍼 캠벨과 폴 그리스, 제이슨 몬토조, 그렉 윌슨 공저 'Practical Programming' P.137에서 말 함.

## 2. dictionary(딕셔너리)

#### ➥ 예제: 다음 한 사람에 관한 정보를 갖는 JSON객체이다.
#### 키-값 쌍(이름:값)으 패턴으로 표현된다.

```python
a = {'이름: '홍길동',
    '나이': 25,
    '성별': '여',
    '주소': "서울 특별시 양천구 목동",
    '특기': ['농구','도술'],
    '가족관계': {'#': 2, '아버지': '홍판서', '어머니': '춘섬'}
    '회사': '경기 수원시 팔당구 우만동'
}
```

#### 딕셔너리를 *연관배열(Associative array) 또는 해시(Hash)라고도 부른다.* Key를 통해 Value를 얻는다.

```python
dic = {'name': 'Eric', 'age':15}
print(dic['name'])
print(dic['age'])
```

#### ➥ *list 같은 경우 print(dic[0]) 이런식으로 순서를 통해 값을 불러 왔지만* *dictionary의 경우 key를 통해 값을 불러올 수 있다.*

#### 1.딕셔너리 쌍 추가하기
#### 🔥주의. 추가와 수정의 문법이 같다: 딕셔너리[키] = 값 형태로 코드를 작성할 때,
#### ① 없는 키를 넣으면 추가가 되고 ② 이미 있는 키를 넣으면 기존 값이 수정(덮어쓰기) 된다.

```python
a = {1:'a'}
a[2] = 'b'
print(a)
```

```python
a = {1:'a'}
a[1] = 'b'
print(a)
```

#### 👍🏻 *List와는 달리 dictionary는 추가 하는 방법이 다르다.*
#### 아래와 같이 *① 좌축은 Key를, ② 우측은 Value* 를 배치하면 dictionary에 추가 된다.

```python
a = {1:'a'}
a['name'] = '익명'

print(a)
```

#### 2. 딕셔너리 요소 삭제하기
##### *딕셔너리에는 순서가 없고 Key와 value만 있을 뿐이다.*
##### *따라서 List 처럼 a[0] 이런게 필요 없다.*

##### ⤋ 아래 예시에서 *del a[1] 만 해도 {1,:'a'}는 사라진다* 는 걸 보여준다. 즉 키와 값은 쌍으로 함께 움직인다.

```python
a = {1:'a'}
a['name'] = '익명'

print(a)

del a[1] #key를 입력하면 쌍으로 지워진다.

print(a)
```

#### 3. Key를 이용해서 Value 얻기

```python
grade = {'pey': 10, 'julliet': 99}
print(grade['pey'])
```

#### 4. Dictionary 만들 때 🔥 주의 사항: *"Value는 겹칠 수 있지만 KEY는 중복되어서는 절대 안 된다."*

```python
a= {1:'a', 1:'b', 3:'c'}
print(a)
```

```python
a= {1:'a', 2:'a', 3:'c'}
print(a)
```

#### 5. 🔥 *Key 와 Value를 List처럼 만들기(keys) / 🔥 .items()의 경우 Tuple 형태로 만들어 진다.*

```python
a = {'name':'pey','phone':'010-999-9999','birth': '11.08'}
print(a.keys())
print(a.values())
print(a.items())

for k in a.keys():  # 마치 a의 list인 것처럼 만드는 반복문으로 사용될 수 있다.
    print(k)

for key, value in a.items():  # items()를 사용하면 key와 value를 쌍으로 꺼낼 수 있습니다.
    print(f"키는 {key}이고, 값은 {value}입니다.")

print(a.get(4))
print(a.get(4,'없음'))
print('name' in a)
print(4 in a)
print(a.clear())
```

## 3.&nbsp;set(집합)

#### 🔥집합의 핵심은 *List* 와 달리🌓 **'중복된 값'** 을 넣지 못한다는 것이다. 또한 집합의 원소는 *List* 와 달리 순서가 없다.

```python
s1 = set([1,2,3])
s1 = {1,2,3} # 둘다 모두 집함의 원소를 돌려준다.
print(s1)
```

##### *집합은 데이터를 다루면 중복되는 걸 다루는 데 많이 쓰인다. 중복 값을 '찾거나', 중복 값을 '빼는 것'과 같은 것들.*

```python
y= [1,2,3,3,3,4,4,5,6,7,7,7]
newlist = set(y) #중복값을 넣지 못한다는 집합의 성질을 활용한 것이다.
print(newlist)
x = [1,2,3,3,3,4,4,5,6,7,7,7]
newlist = list(set(x)) #집합의 성질을 이용해 새로운 리스트를 만든 것이다.
print(newlist)
```

#### 교집합 1

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7.8,9])
print(s1 & s2)
```

#### 교집합 2

```python
print(s1.intersection(s2))
```

#### 합집합 1

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7.8,9])
print(s1 | s2)
```

#### 합집합 2

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7.8,9])
print(s1.union(s2))
```

#### 차집합 1

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7.8,9])

print(s1-s2)
```

#### 차집합 2

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7.8,9])

print(s1.difference(s2))
print(s2 - s1)
print(s2.difference(s1))
```

#### 집합에서 값 1개 추가하기(add)

```python
s1 = set([1,2,3,4,5,6])

s1.add(8)
print(s1)
```

#### 집합에서 값 여러 개 추가하기(update)

```python
s1 = set([1,2,3,4,5,6])
s1.update([7,8,9,10])  # 여러개를 추가할 때
print(s1)
s1.add(11) # 한 개 추가할 때
print(s1)
```

#### 집합에서 특정 값 제거하기(remove)

```python
s1 = set([1,2,3,4,5,6])
s1.remove(3)
print(s1)
```

#### 4.&nbsp; Boolen(불)

##### 불은 참(true), 거짓(false)을 판별하는 걸 말한다.

```python
a = True # True 자체가 자료형이다. 그래서 'True' 이렇게 쓰지 않는다. 그리고 반드시 대문자 T를 써야한다.
print(a)
print(type(a))
```

### 🔥자료형의 참과 거짓
- #### "python"   참
- #### " "        거짓
- #### [1,2,3]    참
- #### []         거짓
- #### ()         거짓
- #### {}         거짓
- #### 1          참
- #### 0          거짓
- #### None       거짓
##### 안에 문자가 채워져 있으면 참이고 빈 문자열이면 거짓이다. 특이하게 숫자 1은 참이고 0은 거짓이다.

```python
a = True
if a:
    print(a)
```

#### 아래의 경우 문자열 안에 무언가 채워져 있으므로 True로 판단한다.

```python
a = "안녕하세요"
if a: # True인지 False인지를 판단하는 조건문 if
    print(a)
```

#### 아래의 경우 문자열 안에 무언가 비워져 있으므로 False 로 판단한다.

```python
a = "  "
if a:
    print(a) # 이 경우 값이 아예 안 나온다. 즉 "  "를 거짓으로 본 것이다.
```

```python
a = [1,2,3,4]
if a:
    print(a)
```

```python
a = [1,2,3,4]
while a:
    a.pop() # pop은 마지막 자리에 놓인 값을 없앤 후 반환한다. 여기는 반복문이니 값이 []될 떄까지 반복한다. 왜냐하면 a에 값이 없어졌을 때는 a.pop()의 결과가 False가 되므로 인식하기 때문이다.
    print(a)
```
