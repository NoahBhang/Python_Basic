
| 파일 읽기 모드 | 설명 |
|:---------| :--- |
| **r**    | 읽기 모드 - 파일을 읽기만 할 때 사용 |
| **w**    | 쓰기 모드 - 파일에 내용을 쓸 때 사용 |

### 1-2 쓰기모드 예시
- ##### *w (쓰기 모드) 예시&nbsp;특징: 파일이 없으면 새로 만들고, 이미 있으면 기존 내용을 모두 지우고 새로 쓴다.*
- ##### *파이썬에서 파일을 열고 닫을 때는 open(), close() 함수를 사용한다.*

````python
# test.txt 파일을 쓰기 모드로 열기
f = open("test.txt", "w", encoding="utf-8")

# 파일에 내용 작성하기
f.write("안녕하세요.\n")
f.write("파이썬 파일 쓰기 연습.\n")

# 파일 닫기 (반드시 닫아줘야 저장된다)
f.close()
````

### 1-3 &nbsp; 파일을 쓰기 모드로 열어 출력하기

#### 1) &nbsp;절대 주소: 처음 부분(c:/)부터 주소를 써주는 것.
#### 2) &nbsp; 상대 주소: 현재 실행한는 파일을 기준으로 상대적인 경로를 써주는 것.

#### 🔥맥을 사용할 때와 윈도우 사용할 때 다름을 주의하자.

````python
# 아래의 경우는 윈도우를 사용 할 때이다.
# c:/python/새파일.txt 전체를 따옴표로 감싸주어야 한다.
f = open("c:/python/새파일.txt", 'w', encoding='utf-8')

for i in range(1, 11):
    data = "%d번째 줄 이다.\n" % i
    f.write(data)

f.close()
````

# 파일 입출력(Read/Write) 문법의 현업 활용 요약

기초 문법책의 지루한 숫자 예제와 달리, 파일 입출력은 프로그램이 종료되어도 데이터를 영구적으로 저장하고 다시 불러올 수 있는 유일한 수단이다. 현업 개발자들이 출근해서 이 문법을 실제로 사용하는 대표적인 상황들을 정리했다.

---

## 1. 일반적인 현업 활용 예시

### ☕ 대량 데이터 일괄 처리 (Batch Processing)
* **상황**: 마케팅팀에서 엑셀 파일(`.csv`)로 회원 10만 명의 이메일 주소를 주며 커스텀 쿠폰 메시지 생성을 요청함.
* **활용**: `open("members.csv", "r")`로 회원 명단을 읽어와 반복문을 돌리면서, 개별 회원의 이름을 넣은 커스텀 메시지를 `open("coupon_msg.txt", "w")`로 단 몇 초 만에 대량 생성함.

### 🚨 서버 로그 파일 분석 및 트러블슈팅
* **상황**: 잘 돌아가던 웹사이트가 새벽에 갑자기 멈췄으나 화면에는 아무런 단서도 남지 않음.
* **활용**: 프로그램의 모든 이벤트가 기록되는 `log.txt` 파일을 `open("log.txt", "r")`로 읽어와 특정 에러 키워드(`[ERROR]`, `[CRITICAL]`)가 포함된 줄만 골라내어 버그의 원인을 역추적함.

### 📊 주기적인 데이터 누적 관리
* **상황**: 매일 오후 3시 30분에 주식 시장이 마감되면 수집된 당일 주가 데이터를 내일 분석할 수 있도록 기록해 두어야 함.
* **활용**: 수집한 당일 주가 정보를 `open("stock_history.txt", "a")` 추가 모드를 사용하여 기존 데이터 파일의 맨 아래에 한 줄씩 이어 붙여 누적함.

### 🎮 프로그램 상태 저장 (Save & Load)
* **상황**: 사용자가 모바일 앱이나 게임을 하다가 도중에 종료했다가 다시 실행함.
* **활용**: 종료 직전의 상태(레벨, 장바구니, 설정 등)를 `open("save_data.json", "w")`로 파일에 기록하고, 다시 앱을 켰을 때 `open("save_data.json", "r")`로 읽어와 마지막 화면을 그대로 복원함.

## 2. 쿠팡 로켓연구소 FDS(이상금융거래탐지시스템)팀에서의 활용

쿠팡 FDS팀은 수천만 건의 결제 데이터와 회원 행동 패턴을 실시간 또는 주기적으로 분석하여 부정 거래(아이디 도용, 대포통장 결제, 매크로 주문 등)를 잡아낸다.

* **FDS 룰 및 블랙리스트 동기화 (Read)**
  * 부정 거래자로 의심되거나 이미 차단된 위험 IP 리스트, 사기 패턴 룰셋(Rule-set) 파일들을 `open("blacklist.json", "r")` 구조로 읽어와 메모리에 실시간 탐지 엔진으로 탑재한다.
* **탐지된 부정 결제 로그 영구 저장 (Write)**
  * 실시간 결제 요청 중 사기 거래 확률이 높은 위험 거래가 적발되면, 해당 유저의 ID, IP, 결제 금액, 사유 등을 `open("fds_alerts.log", "a")` 추가 모드로 파일에 신속하게 기록한다. 이 파일은 향후 수사 기관 제출용 증거 자료 및 빅데이터 분석의 기초가 된다.

  ```python

      transactions = f.readlines()
  ```

### 🔍 2단계: 이상 거래(재산은닉/편파변제) 의심 내역 선별하기

### 💾 3단계: 파산관재인 제출용 ' FDS 의심 거래 보고서' 생성 (Write)
* **상황**: 분석 엔진이 찾아낸 이상 거래 내역들만 모아서 파산관재인이 법원에 제출할 수 있도록 별도의 텍스트 파일로 자동 저장함.
* **코드 적용**:

  ```python
      for alert in fds_alerts:
          f.write(alert + "\n")
  ```

먼저 분석할 대상인 원본 데이터 파일을 만든다. 주피터 노트북 셀에 아래 코드를 넣고 실행하면 현재 폴더에 `raw_bank_statement.txt` 파일이 생성된다.

```python
with open("raw_bank_statement.txt", "w", encoding="utf-8") as f:

    f.write("2026-02-01,마트결제,45000,일반\n")
```

### 🔍 2단계: 파일 읽기 및 특정 조건(500만 원 이상) 데이터 추출·저장
생성된 원본 파일을 읽어와 한 줄씩 분석한 뒤, 조건에 맞는 데이터만 `fds_suspicious_report.txt` 파일로 선별 저장하는 핵심 FDS 로직이다.

```python
# 1. 원본 데이터 읽기 모드로 열기
with open("raw_bank_statement.txt", "r", encoding="utf-8") as raw_file:
    # 2. 결과 보고서를 작성할 쓰기 모드로 열기
    with open("fds_suspicious_report.txt", "w", encoding="utf-8") as report_file:

        # 보고서 상단 헤더 작성
        report_file.write("=== 파산자 고액 의심 거래 내역 보고서 ===\n")
        report_file.write("기준: 5,000,000원 이상 거래 건\n")
        report_file.write("-" * 40 + "\n")

        # 원본 데이터를 한 줄씩 읽어가며 조건 검사
        for line in raw_file:
            # 쉼표(,)를 기준으로 데이터 분리 [날짜, 적요, 금액, 유형]
            data = line.strip().split(",")

            # 금액 데이터(인덱스 2번)를 정수형(int)으로 변환
            amount = int(data[2])

            # FDS 탐지 조건: 금액이 5,000,000원 이상인 경우
            if amount >= 5000000:
                # 탐지된 라인을 보고서 파일에 그대로 기록
                report_file.write(f"[⚠️ 탐지] 날짜: {data[0]} | 내역: {data[1]} | 금액: {amount:,}원\n")

print("FDS 의심 거래 분석 및 보고서 생성이 완료되었다.")
```

    ```text
    [⚠️ 탐지] 날짜: 2026-02-12 | 내역: 현금인출 | 금액: 6,000,000원
    ```

# 파산자 FDS를 위한 내장 csv 라이브러리 정교하게 다루기

실제 금융권이나 쿠팡 FDS팀이 다루는 거래 내역 데이터는 대부분 쉼표로 구분된 **CSV(Comma-Separated Values)** 형식이다. 앞서 사용한 `split(",")` 방식은 데이터 내부에 쉼표가 포함되어 있거나 문장 부호가 섞여 있으면 데이터가 완전히 깨지는 치명적인 한계가 있다.

파이썬 내장 `csv` 라이브러리를 사용하면 복잡한 금융 데이터를 에러 없이 안전하고 정교하게 처리할 수 있다. 주피터 노트북에 붙여넣어 바로 실습할 수 있는 코드 가이드이다.

---

## 1. 일반 split(",") 방식의 치명적인 한계

만약 파산자가 재산을 은닉하기 위해 아래와 같이 적요란에 쉼표를 넣어 송금했다고 가정하자.
> `2026-01-15,"지인송금, 차명계좌로분할",7500000`

이 데이터를 기존 방식인 `line.split(",")`으로 쪼개면 다음과 같이 쪼개진다.
* `data[0]`: `2026-01-15`
* `data[1]`: `"지인송금`
* `data[2]`: `차명계좌로분할"`
* `data[3]`: `7500000`

금액 데이터가 2번 인덱스가 아닌 3번 인덱스로 밀려나기 때문에 `int(data[2])`를 실행하는 순간 프로그램이 통째로 뻗어버리는 데이터 오염이 발생한다. **내장 `csv` 라이브러리는 큰따옴표 안의 쉼표를 데이터로 인식하여 이 문제를 완벽하게 해결해 준다.**

## 2. csv 라이브러리를 활용한 정교한 읽기 및 쓰기 실습

### 🛠️ 1단계: 실습용 고도화된 가상 CSV 파일 생성
큰따옴표와 쉼표가 섞여 있는 실제 금융 데이터와 유사한 환경을 만든다.

```python
import csv

# 실제 프로젝트 환경처럼 컬럼명(Header)이 포함된 CSV 파일 생성
with open("advanced_bank_statement.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)

    # 1. 컬럼 이름 정의 (헤더)
    writer.writerow(["거래일자", "적요", "거래금액", "계좌번호"])

    # 2. 데이터 작성 (적요 내부에 쉼표가 들어가도 csv 라이브러리가 자동으로 처리함)
    writer.writerow(["2026-01-10", "마트 결제", "35000", "110-123-456"])
    writer.writerow(["2026-01-15", "지인송금, 차명계좌분할", "7500000", "110-999-888"]) # 위험 거래
    writer.writerow(["2026-02-01", "현금 인출, ATM기기", "6000000", "110-123-456"])  # 위험 거래
    writer.writerow(["2026-02-12", "월세 납부", "500000", "110-555-444"])
```

### 🔍 2단계: csv.reader를 이용한 데이터 파싱 및 의심 거래 필터링
생성된 CSV 파일을 안전하게 읽어와 첫 줄(헤더)을 건너뛰고 금액을 추출하는 FDS 실무 로직이다.

```python
import csv

# 추출된 의심 거래를 담을 리스트
suspicious_transactions = []

with open("advanced_bank_statement.csv", "r", encoding="utf-8") as f:
    # csv.reader 객체 생성
    reader = csv.reader(f)

    # 첫 번째 줄(컬럼명 헤더)은 데이터 분석에서 제외하기 위해 건너뜀
    header = next(reader)

    for row in reader:
        # csv 라이브러리가 큰따옴표 안의 쉼표를 정확히 처리해주므로
        # 거래금액은 항상 정확하게 인덱스 2번에 위치함
        trade_name = row[1]
        amount = int(row[2])
        account = row[3]

        # FDS 탐지 조건: 500만 원 이상 고액 거래
        if amount >= 5000000:
            suspicious_transactions.append(row)
            print(f"[🚨 탐지] 적요: '{trade_name}' | 금액: {amount:,}원 | 계좌: {account}")
```

### 💾 3단계: 탐지된 결과를 컬럼명을 유지한 채 새 CSV 파일로 저장
관재인이 엑셀로 바로 열어볼 수 있도록 헤더를 포함한 독립된 CSV 파일로 결과물을 출력한다.

```python
# 탐지된 위험 거래만 모아서 fds_csv_report.csv 파일로 깔끔하게 저장
with open("fds_csv_report.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)

    # 보고서 파일에도 컬럼 이름(헤더)을 먼저 작성해줌
    writer.writerow(["거래일자", "이상적요내역", "위험금액", "대상계좌"])

    # 탐지된 모든 행(List)을 한 번에 파일에 씀
    writer.writerows(suspicious_transactions)

print("\n관재인 제출용 이상거래 CSV 보고서 생성이 완료되었다.")
```

## 3. 실무 효율을 극대화하는 꿀팁: `csv.DictReader`

인덱스 번호(`row[1]`, `row[2]`)로 데이터를 가져오면 나중에 CSV 파일의 컬럼 순서가 바뀔 때 코드를 전부 수정해야 하는 대참사가 일어난다. 이때 컬럼 이름으로 직접 데이터를 가져오는 `DictReader`를 쓰면 코드가 훨씬 직관적이고 안전해진다.

```python
import csv

with open("advanced_bank_statement.csv", "r", encoding="utf-8") as f:
    # 데이터를 파이썬 딕셔너리(Dictionary) 형태로 읽어옴
    reader = csv.DictReader(f)

    for row in reader:
        # 인덱스 번호 대신 엑셀 상단의 '컬럼 이름'을 키값으로 바로 접근함
        # 컬럼 순서가 바뀌어도 코드가 깨지지 않아 FDS 실무에서 애용됨
        amount = int(row["거래금액"])
        if amount >= 5000000:
            print(f"[딕셔너리 탐지] 대상계좌: {row['계좌번호']} -> {amount:,}원 은닉 의심")
```

# 금융 데이터 FDS 분석을 위한 필수 문자열 및 공백 전처리 기법

실제 은행이나 카드사에서 내려받는 거래 내역서(CSV, TXT)는 사람이 손으로 입력하거나 시스템 간 연동 과정에서 수많은 공백, 특수문자, 대소문자 불일치 등의 오염이 발생한다.

예를 들어 파산자가 추적을 피하기 위해 적요란에 `"홍 길 동 "`, `" 홍길동"`, `"홍길동(가족)"`처럼 꼼수를 부려두면, 컴퓨터는 이를 모두 다른 사람으로 인식하여 추적망을 놓치게 된다. 따라서 파일에서 데이터를 읽어온 직후 **정밀한 전처리(Preprocessing)** 를 수행하는 것은 FDS 엔진 구축의 핵심 필수 단계이다.

---

## 1. 금융 데이터 오염의 대표적인 사례와 전처리 목표

* **사례 1 (불규칙한 공백)**: " 홍길동", "홍 길동", "홍길동 " → 모두 동일한 인물 "홍길동"으로 일치시켜야 함.
* **사례 2 (숨겨진 제어문자)**: 눈에 보이지 않는 줄바꿈(`\n`)이나 탭(`\t`) 기호가 문자열 앞뒤에 붙어 금액 변환 오류 유발 → 완벽히 제거해야 함.
* **사례 3 (금액 표기 오염)**: "5,000,000원" 또는 " 5000000 " → 순수한 숫자 형태의 문자열 "5000000"로 정제 후 `int`형 변환해야 함.
* **사례 4 (영문 대소문자 혼재)**: 계좌주 이름이나 기관명이 "Coupang", "COUPANG", "coupang"으로 제각각 → 하나로 통일해야 함.

---

## 2. 파이썬 내장 기능을 활용한 핵심 전처리 실습

주피터 노트북에 아래 코드를 복사하여 실행하면, 심각하게 오염된 금융 데이터를 FDS 탐지가 가능한 깨끗한 데이터로 정제하는 과정을 확인할 수 있다.

```python
import csv
```

```python
# 2. 오염된 파일을 읽어와 정밀 전처리 프로세스 적용하기
with open("dirty_bank_statement.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    print(f"{'[' + '전처리 전' + ']':<25} ➡️  {'[' + '전처리 후 FDS 탐지 상태' + ']':<25}")
    print("-" * 75)

    for row in reader:
        # --- [기법 1] 문자열 앞뒤 공백 및 숨은 제어문자(\n, \t) 완전 제거: strip() ---
        raw_name = row["적요(수취인)"]
        clean_name = raw_name.strip()

        # --- [기법 2] 글자 사이에 숨은 모든 악의적 공백 제거: replace() ---
        clean_name = clean_name.replace(" ", "")

        # --- [기법 3] 영문 데이터 일관성을 위한 대문자 통일: upper() (또는 lower()) ---
        clean_name = clean_name.upper()

        # --- [기법 4] 금액 필드의 특수문자(쉼표, '원' 글자) 정제 및 공백 지우기 ---
        raw_amount = row["거래금액"]
        clean_amount = raw_amount.replace(",", "").replace("원", "").strip()

        # 최종 수치형 데이터 변환 (전처리가 정상적으로 끝나야 에러가 안 남)
        amount_int = int(clean_amount)

        # FDS 엔진 탐지 기준 적용 (예: 500만 원 이상 위험 거래)
        status = "🟢 정상"
        if amount_int >= 5000000:
            status = "🚨 고액 은닉 의심"

        print(f"이름: {raw_name:<10} 금액: {raw_amount:<12} ➡️  이름: {clean_name:<10} 금액: {amount_int:<10,}원 [{status}]")
```

# 파산자 FDS를 위한 정규표현식(re) 금융 데이터 적용 가이드

금융 거래 내역서의 적요란은 사용자가 자유롭게 입력하기 때문에 문자열 패턴이 무작위적이다. 예를 들어 파산자가 친인척에게 돈을 빼돌리면서 `"아들송금"`, `"자녀(홍길동)"`, `"차남 학비"`, `"정기_지원_금"` 처럼 변칙적으로 기록해 두면 일반적인 `in` 연산자나 `replace()`만으로는 전부 골라내기 어렵다.

파이썬 내장 **정규표현식(re) 라이브러리**를 사용하면 이처럼 불규칙하게 오염된 텍스트 속에서 특정 '패턴'을 정교하게 컴퓨터에게 학습시켜 솎아낼 수 있다. 주피터 노트북에 바로 붙여넣어 실행할 수 있는 실무 예제이다.

## 1. FDS 현업에서 정규표현식을 쓰는 결정적 이유

* **기존 방식의 한계**: `if "아들" in text or "딸" in text or "자녀" in text:` ➞ 우회 키워드가 나올 때마다 if 문이 무한정 길어지고 예외 처리가 불가능하다.
* **정규표현식 방식**: `[아딸자녀|차남|장녀]` 라는 하나의 패턴 템플릿을 정의하여, 단 한 줄의 코드로 수십 가지 우회 패턴을 통제한다.

```python
family_pattern = re.compile(r".*(장녀|차남|아들|딸|자녀).*")

 패턴 2: 텍스트 내부에 숨겨진 계좌번호 형태 탐지
 숫자 3~6자리 - 숫자 2~4자리 - 숫자 3~6자리 패턴 (\d는 숫자를 의미)
account_pattern = re.compile(r"\d{3,6}-\d{2,4}-\d{3,6}")

        if family_pattern.search(memo):

        account_match = account_pattern.search(memo)
        if account_match:
```

파이썬의 **사용자 정의 함수(`def`)** 를 활용하여 데이터 정제 로직을 하나의 '공정 라인(파이프라인)'처럼 묶어두면, 어떤 지저분한 데이터가 들어와도 함수 한 줄로 완벽하게 전처리할 수 있다. 주피터 노트북용 실무 코드이다.

## 1. FDS 전처리 전용 사용자 정의 함수 설계

FDS 엔진에 입력하기 전, 적요와 금액 데이터를 완벽하게 세척해 주는 핵심 함수 2개를 정의한다.

```python
import re

def clean_transaction_memo(memo):
    """
    적요란의 텍스트 오염을 제거하고 일관된 포맷으로 변경하는 함수
    - 앞뒤 공백 및 글자 사이 공백 완전 제거
    - 영문 대문자 통일
    """
    if not memo:
        return ""

    1. 문자열 변환 및 앞뒤 공백 제거
    cleaned = str(memo).strip()
    2. 글자 사이의 모든 공백 제거
    cleaned = cleaned.replace(" ", "")
    3. 영문 대문자화
    cleaned = cleaned.upper()

    return cleaned


def clean_transaction_amount(amount_text):
    """
    금액 텍스트 속 특수문자(쉼표, 원, 공백 등)를 제거하고
    FDS 연산이 가능한 순수 정수(int)형으로 변환하는 함수
    """
    try:
        # 정규표현식을 이용해 숫자가 아닌 모든 문자([^0-9])를 제거
        cleaned_numbers = re.sub(r"[^0-9]", "", str(amount_text))

        if cleaned_numbers == "":
            return 0

        return int(cleaned_numbers)

    except (ValueError, TypeError):
        # 데이터가 손상되어 숫자로 바꿀 수 없는 경우 예외 처리
        return 0
```

# 온프레미스 배포용 단일 파일 룰(Rule) 기반 Pandas FDS 분석 가이드

이 코드는 복잡한 모듈화나 외부 API 연동 없이, **파이썬과 Pandas 라이브러리만 설치된 오프라인 내부 서버(On-Premise)** 에서 단 하나의 주피터 노트북 파일로 완벽하게 구동되도록 설계되었다.

머신러닝 전 단계에서 가장 확실하고 강력한 **'고정 규칙(Rule-based)'** 방식을 적용하여 파산자의 은닉 의심 거래를 대량으로 통제한다.

---

## 1. 관재인용 핵심 FDS 탐지 룰(Rule) 설정
본 프로그램은 파산관재인 실무에서 가장 유의 깊게 보는 3가지 핵심 규칙을 기반으로 움직인다.
*   **Rule 1**: 500만 원 이상의 고액 거래 (재산 은닉 의심)
*   **Rule 2**: 적요란에 가족 관련 우회 키워드 포함 (편파 변제 및 은닉 의심)
*   **Rule 3**: 파산 직전(예: 2026년 3월 이후)에 발생한 집중 거래

---

## 2. Pandas 기반 단일 파일 핵심 FDS 소스 코드

주피터 노트북의 단 하나의 셀에 아래 코드를 통째로 붙여넣고 실행하면 데이터 적재, 전처리, 룰 필터링, 결과 파일 출력까지 일사천리로 진행된다.

```python
import pandas as pd
import numpy as np

# ==========================================
# [STEP 1] 온프레미스 가상 금융 데이터셋 생성
# ==========================================
# 실습을 위해 컬럼과 오염된 데이터가 섞인 거래내역을 즉석에서 생성한다.
raw_data = {
    "거래일자": ["2026-01-10", "2026-02-15", "2026-03-02", "2026-03-12", "2026-04-05"],
    "적요": [" 마트 삼겹살 ", "지인 송금(홍길동)", " 차남 학자금 지원 ", "개인 차량 주유비", " 와이프 현금전달 "],
    "거래금액": ["45,000", "7,500,000 원", "12,000,000", "70,000", "8,500,000"]
}
df = pd.DataFrame(raw_data)
print("=== [성공] 온프레미스 원본 데이터 로드 완료 ===")
print(df, "\n")


# ==========================================
# [STEP 2] Pandas 초고속 일괄 전처리 (No for-loop)
# ==========================================
# 반복문을 돌리지 않고 수만 행의 데이터를 한 번에 세척한다.

# 1. 적요란의 모든 공백 제거 및 대문자화
df["정제적요"] = df["적요"].str.replace(" ", "").str.upper()

# 2. 거래금액에서 숫자가 아닌 문자(쉼표, 원)를 한 번에 제거하고 정수형(int) 변환
df["정제금액"] = df["거래금액"].str.replace(r"[^\d]", "", regex=True).astype(int)


# ==========================================
# [STEP 3] 룰 기반(Rule-based) FDS 탐지 매칭
# ==========================================
# 관재인이 지정한 룰 조건을 정의한다.

# 조건 A: 500만 원 이상 고액 거래
cond_amount = df["정제금액"] >= 5000000

# 조건 B: 위험 키워드(차남, 와이프, 지인, 현금) 포함 여부 검사
danger_keywords = "차남|와이프|지인|현금|가족"
cond_keyword = df["정제적요"].str.contains(danger_keywords)

# 조건 C: 파산 신청 시점인 2026년 3월 이후 거래
cond_date = df["거래일자"] >= "2026-03-01"

# 정의한 룰(조건)들을 결합하여 최종 '이상 거래' 유무 컬럼을 생성한다.
# 금액이 500만 원 이상이면서 + (위험 키워드가 있거나 or 파산 직전인 경우)
fds_rule = cond_amount & (cond_keyword | cond_date)

# 기본값은 '정상'으로 두고, 룰에 걸린 데이터만 '🚨 고위험 은닉 의심'으로 분류
df["FDS결과"] = np.where(fds_rule, "🚨 고위험 은닉 의심", "🟢 정상")


# ==========================================
# [STEP 4] 관재인 제출용 최종 보고서 파일 저장
# ==========================================
# 룰에 탐지된 위험 행들만 쏙 골라내어 데이터프레임을 분리한다.
fds_report_df = df[df["FDS결과"] == "🚨 고위험 은닉 의심"]

# 필요한 컬럼만 보기 좋게 정리
fds_report_df = fds_report_df[["거래일자", "적요", "정제금액", "FDS결과"]]

# 온프레미스 환경에서 엑셀이나 관재인 시스템이 바로 읽을 수 있게 CSV로 출력
fds_report_df.to_csv("fds_bankruptcy_final.csv", index=False, encoding="utf-8-sig")

print("=== [완료] 룰 기반 FDS 탐지 완료 및 보고서 저장 완료 ===")
print(fds_report_df)
```

### 💡 온프레미스 가벼운 배포를 위한 Pandas 팁
### encoding="utf-8-sig" 활용: 파일 저장 시 일반 utf-8 대신 utf-8-sig를 사용하면, 인터넷이 안 되는 관재인의 PC에서 엑셀(Excel)로 이 CSV 파일을 더블클릭해 열어도 한글이 절대 깨지지 않는다.
### 가독성 최우선: 함수를 쪼개서 파일 여러 개로 만들면 배포할 때 경로가 꼬이기 쉽지만, 이렇게 하나의 주피터 파일 안에서 위에서 아래로 흐르게 짜면 다른 오프라인 PC에 가져가서 실행해도 에러 없이 똑같이 돌아간다.

# 온프레미스 단일 파일용 대화형 입력창(input) 설정 가이드

온프레미스 환경에서 별도의 웹 인터페이스(GUI)를 구축하지 않고도, 파이썬 내장 함수인 `input()`을 활용하면 파산관재인이 주피터 노트북 셀을 실행할 때마다 **원하는 탐지 기준 금액과 위험 키워드를 직접 입력하여 커스텀 보고서를 뽑아낼 수 있다.**

모듈화 없이 하나의 스크립트 안에서 관재인의 입력에 따라 동적으로 FDS 룰이 바뀌도록 코드를 고도화했다.

---

## 1. 대화형 input() 적용 FDS 통합 소스 코드

주피터 노트북 셀에 아래 코드를 붙여넣고 실행하면, 콘솔창에 금액과 키워드를 묻는 입력창이 차례로 나타난다. 관재인이 수치를 바꾸는 대로 즉시 결과 파일에 반영된다.

```python
import pandas as pd
import numpy as np

# ==========================================
# [STEP 1] 기본 금융 데이터셋 준비
# ==========================================
raw_data = {
    "거래일자": ["2026-01-10", "2026-02-15", "2026-03-02", "2026-03-12", "2026-04-05"],
    "적요": [" 마트 삼겹살 ", "지인 송금(홍길동)", " 차남 학자금 지원 ", "개인 차량 주유비", " 와이프 현금전달 "],
    "거래금액": ["45,000", "3,500,000", "12,000,000", "70,000", "8,500,000"]
}
df = pd.DataFrame(raw_data)

# 데이터프레임 기본 전처리 (공백 제거 및 숫자형 변환)
df["정제적요"] = df["적요"].str.replace(" ", "").str.upper()
df["정제금액"] = df["거래금액"].str.replace(r"[^\d]", "", regex=True).astype(int)


# ==========================================
# [STEP 2] 관재인 대화형 입력창 (Interactive Input)
# ==========================================
print("============ 파산관재인 FDS 조건 설정 ============")

# 1. 기준 금액 입력 받기 (예: 3000000)
# 사용자가 입력한 문자열을 정수(int)로 변환해야 조건 연산이 가능하다.
try:
    user_limit = input("👉 탐지할 기준 금액을 입력하세요 (예: 3000000): ")
    target_amount = int(user_limit)
except ValueError:
    print("[경고] 올바른 숫자가 입력되지 않아 기본값인 5,000,000원으로 설정한다.")
    target_amount = 5000000

# 2. 탐지 키워드 입력 받기 (예: 차남, 와이프)
# 관재인이 쉼표로 구분해 입력하면 Pandas용 패턴 기호(|)로 자동 변환한다.
user_keywords = input("👉 탐지할 위험 키워드를 쉼표로 구분해 입력하세요 (예: 차남,와이프,지인): ")
clean_keywords = user_keywords.replace(" ", "").replace(",", "|")

print("\n==================================================")
print(f"🎯 설정된 필터 조건: {target_amount:,}원 이상 / 위험키워드: [{user_keywords}]")
print("==================================================\n")


# ==========================================
# [STEP 3] 동적 룰 기반(Rule-based) FDS 탐지
# ==========================================
# 관재인이 입력한 변수(target_amount, clean_keywords)를 룰에 대입한다.

cond_amount = df["정제금액"] >= target_amount
cond_keyword = df["정제적요"].str.contains(clean_keywords)
cond_date = df["거래일자"] >= "2026-03-01"

# 최종 탐지 규칙 결합
fds_rule = cond_amount & (cond_keyword | cond_date)
df["FDS결과"] = np.where(fds_rule, "🚨 고위험 은닉 의심", "🟢 정상")


# ==========================================
# [STEP 4] 맞춤형 최종 보고서 저장
# ==========================================
fds_report_df = df[df["FDS결과"] == "🚨 고위험 은닉 의심"]
fds_report_df = fds_report_df[["거래일자", "적요", "정제금액", "FDS결과"]]

# 관재인이 구분하기 쉽도록 입력한 금액 조건이 명시된 파일명으로 자동 저장된다.
output_filename = f"fds_report_{target_amount}원_이상.csv"
fds_report_df.to_csv(output_filename, index=False, encoding="utf-8-sig")

print(f"=== [완료] '{output_filename}' 파일이 안전하게 생성되었다. ===")
print(fds_report_df)
```

# 온프레미스용 실제 엑셀(.xlsx) 파일 Pandas 직접 연동 가이드

파산관재인 실무 현장이나 법원 데이터는 CSV보다 보안이 적용된 실제 엑셀 파일(`.xlsx`)로 오가는 경우가 압도적으로 많다. 파산관재인이 은행이나 카드사에서 내려받은 **순수 엑셀 양식을 파일 변환 없이 그대로** 프로그램에 집어넣고, 분석 결과 또한 마케팅이나 법원 제출용 엑셀 서식으로 깔끔하게 내보낼 수 있도록 Pandas의 엑셀 전용 연동 기법을 정리했다.

---

## 1. 온프레미스(오프라인) 환경 설정 시 주의점

인터넷이 연결되지 않는 온프레미스 PC에서 Pandas로 `.xlsx` 파일을 읽고 쓰려면 엑셀 처리용 엔진 라이브러리가 반드시 사전에 설치되어 있어야 한다. 개발 단계나 배포 환경의 터미널(CMD)에서 아래 명령어로 라이브러리를 미리 설치해 둔다.
* **엑셀 읽기 엔진**: `openpyxl` (Pandas 1.0 이상에서 기본 연동)

```bash
pip install openpyxl
```

## 2. 엑셀(.xlsx) 읽기/쓰기 통합 FDS 소스 코드

이 코드는 관재인의 오프라인 PC 폴더 안에 있는 `파산자_은행내역.xlsx` 파일을 읽어와 FDS 규칙으로 필터링한 후, 최종 결과물을 `FDS_관재인_최종보고서.xlsx`라는 실제 엑셀 파일로 출력하는 단일 스크립트 구조이다.

```python
import pandas as pd
import numpy as np
import os

# ==========================================
# [STEP 1] 실습용 가상 엑셀 파일(.xlsx) 즉석 생성
# ==========================================
# 관재인이 외부에서 가져온 엑셀 파일이 이미 폴더에 있다고 가정한 환경을 만든다.
sample_data = {
    "거래일자": ["2026-01-10", "2026-02-15", "2026-03-02", "2026-03-12", "2026-04-05"],
    "적요": [" 마트 삼겹살 ", "지인 송금(홍길동)", " 차남 학자금 지원 ", "개인 차량 주유비", " 와이프 현금전달 "],
    "거래금액": ["45,000", "3,500,000", "12,000,000", "70,000", "8,500,000"]
}
demo_df = pd.DataFrame(sample_data)

# 'openpyxl' 엔진을 사용하여 .xlsx 확장자로 가볍게 저장
demo_df.to_excel("파산자_은행내역.xlsx", index=False, engine="openpyxl")
print("=== [성공] 관재인 입력용 가상 엑셀 파일(파산자_은행내역.xlsx) 생성 완료 ===\n")


# ==========================================
# [STEP 2] Pandas로 실제 엑셀 파일 로드 (Read)
# ==========================================
excel_file = "파산자_은행내역.xlsx"

# 엑셀 파일을 읽어올 때 engine="openpyxl"을 명시하면 오프라인에서도 안전하게 로드된다.
if os.path.exists(excel_file):
    df = pd.read_excel(excel_file, engine="openpyxl")
    print("--- [데이터 확인] 관재인 엑셀 파일 로드 완료 ---")
    print(df, "\n")
else:
    print("[오류] 대상 엑셀 파일이 존재하지 않는다.")


# ==========================================
# [STEP 3] FDS 룰 기반 일괄 데이터 정제 및 매칭
# ==========================================
# 데이터 전처리
df["정제적요"] = df["적요"].str.replace(" ", "").str.upper()
df["정제금액"] = df["거래금액"].str.replace(r"[^\d]", "", regex=True).astype(int)

# 관재인 설정 탐지 룰 규칙 적용 (예: 300만 원 이상 고액 거래 & 위험 키워드)
target_amount = 3000000
danger_keywords = "차남|와이프|지인|현금|가족"

cond_amount = df["정제금액"] >= target_amount
cond_keyword = df["정제적요"].str.contains(danger_keywords)

# 최종 위험 분류
df["FDS결과"] = np.where(cond_amount & cond_keyword, "🚨 고위험 은닉 의심", "🟢 정상")


# ==========================================
# [STEP 4] 분석 결과를 실제 엑셀 파일로 출력 (Write)
# ==========================================
# 룰에 탐지된 위험 행들만 솎아내어 데이터프레임 분리
fds_report_df = df[df["FDS결과"] == "🚨 고위험 은닉 의심"]
fds_report_df = fds_report_df[["거래일자", "적요", "정제금액", "FDS결과"]]

# 최종 결과 파일명 선언
output_excel = "FDS_관재인_최종보고서.xlsx"

# to_excel을 사용하여 관재인이 바로 활용할 수 있는 순수 엑셀 포맷으로 내보낸다.
# index=False 기호를 주어야 첫 번째 열에 불필요한 번호(0, 1, 2)가 생기지 않는다.
fds_report_df.to_excel(output_excel, index=False, engine="openpyxl")

print(f"=== [완료] 실제 엑셀 결과물 '{output_excel}' 파일이 생성되었다. ===")
print(fds_report_df)
```

# 파산자 FDS 최종 MVP: 설명 가능한 코멘트 및 비식별화 엔진

이 코드는 관재인의 노고를 줄여주기 위해 **이상 징후에 대한 구체적인 법적 의심 사유(코멘트)를 자동으로 작성**해 주며, 나중의 머신러닝 고도화를 대비해 **이름과 계좌번호를 완벽하게 마스킹(비식별 처리)** 하여 오프라인에서 안전하게 작동하도록 설계된 단일 파일 소스 코드이다.

## 1. FDS 최종 프로토타입 구현 코드 (Pandas 기반)

주피터 노트북의 셀 하나에 넣고 구동하며, 완성 시 파이참에서 `.py`로 바로 저장하여 배포할 수 있다.

```python
import pandas as pd
import numpy as np

# ==========================================
# [STEP 1] 은행 및 법원 데이터 융합 (가상 데이터)
# ==========================================
# 은행 거래 내역과 법원 제출용 등기/자산 이전 내역이 섞인 통합 데이터셋
raw_financial_data = {
    "거래일자": ["2026-03-05", "2026-03-12", "2026-04-01", "2026-04-15", "2026-05-02"],
    "고객명": ["홍길동", "홍길동", "홍길동", "홍길동", "홍길동"],
    "계좌번호": ["110-123-456789", "110-123-456789", "110-123-456789", "상가등기이전", "110-123-456789"],
    "적요내역": [" 마트 삼겹살 구매 ", " 자녀 홍이슬 증여 ", " 개인 채무 변제(사촌) ", " 소유권 이전 (배우자) ", " 신한은행 현금 인출 "],
    "거래금액": ["45,000", "15,000,000", "8,000,000", "0", "25,000,000"]
}
df = pd.DataFrame(raw_financial_data)


# ==========================================
# [STEP 2] 내부 온프레미스 초고속 전처리
# ==========================================
df["정제적요"] = df["적요"].str.replace(" ", "")
df["정제금액"] = df["거래금액"].str.replace(r"[^\d]", "", regex=True).astype(int)


# ==========================================
# [STEP 3] 설명 가능한 룰 기반(Rule-based) 코멘트 엔진
# ==========================================
# 관재인이 서류를 읽지 않아도 왜 의심했는지 '이유'를 자동으로 달아주는 가중치 룰셋

conditions = [
    (df["정제금액"] >= 5000000) & (df["정제적요"].str.contains("자녀|아들|딸|증여")),
    (df["정제금액"] >= 5000000) & (df["정제적요"].str.contains("사촌|친척|지인|채무")),
    (df["정제적요"].str.contains("소유권|등기|이전") & df["정제적요"].str.contains("배우자|와이프")),
    (df["정제금액"] >= 20000000) & (df["정제적요"].str.contains("현금|인출"))
]

# 각 룰에 매칭될 때 관재인에게 보여줄 전문 코멘트 문구 정의
comments = [
    "⚠️ [재산은닉 의심] 파산 직전 직계존비속 대상 고액 무상 증여 패턴 감지.",
    "⚠️ [편파변제 의심] 특정 친인척 채권자에 대한 우선 변제 및 자산 우회 은닉 의심.",
    "🚨 [부동산은닉 위험] 파산 신청 전후 배우자 명의의 등기 및 자산 소유권 이전 포착.",
    "🔥 [현금유출 심각] 자산 동결 전 대규모 현금 일시 출금을 통한 유동자산 은닉 위험."
]

# 룰에 걸리지 않은 정상 거래는 '정상' 처리
df["분석코멘트"] = np.select(conditions, comments, default="🟢 정상 거래")


# ==========================================
# [STEP 4] 개인정보 마스킹 및 비식별화 (서버 전송용 준비)
# ==========================================
# 나중에 신종 범죄 패턴 공유를 위해 중앙 서버로 보낼 때 개인정보를 철저히 가린다.

# 1. 이름 마스킹 (홍길동 -> 홍*동)
df["비식별_고객명"] = df["고객명"].str.replace(r"(.{1})(.)(.{1})", r"\1*\3", regex=True)

# 2. 계좌번호 마스킹 (110-123-456789 -> 110-123-******)
# 뒤의 6자리를 별표로 채우는 패턴 정규식 적용
df["비식별_계좌"] = df["계좌번호"].str.replace(r"(\d{3}-\d{3}-)\d{6}", r"\1******", regex=True)


# ==========================================
# [STEP 5] 관재인 화면 출력 및 서버 전송용 파일 분리
# ==========================================
# 1. 관재인 PC에 남길 원본 포함 '최종 마스터 보고서'
관재인_보고서 = df[df["분석코멘트"] != "🟢 정상 거래"][["거래일자", "고객명", "계좌번호", "적요내역", "거래금액", "분석코멘트"]]
관재인_보고서.to_excel("관재인_법원제출용_보고서.xlsx", index=False, engine="openpyxl")

# 2. 향후 머신러닝 고도화를 위해 패턴 분석 서버로 내보낼 '비식별 중앙 로그 파일'
서버전송_로그 = df[df["분석코멘트"] != "🟢 정상 거래"][["거래일자", "비식별_고객명", "비식별_계좌", "정제적요", "정제금액", "분석코멘트"]]
서버전송_로그.to_csv("server_fds_pattern_log.csv", index=False, encoding="utf-8-sig")

print("FDS 엔진 분석 및 온프레미스 결과 파일 생성이 안전하게 처리 완료되었다.\n")
print("--- [관재인 화면 노출 대시보드 미리보기] ---")
print(관재인_보고서[["적요내역", "분석코멘트"]])
```

# 제로-리스크(Zero-Risk) 온프레미스 FDS 데이터 전송 엔진

이 코드는 파산관재인 PC 내부(On-Premise)에서 모든 개인정보와 텍스트를 원천 차단(Drop)한다. 외부 본사 서버에는 오직 **'통계성 숫자'** 와 **'추적용 고유 해시 ID(식별 불가능한 난수)'** 만 전송하여, 추후 서버 데이터가 해킹당하더라도 누구의 데이터인지 지구상 그 누구도 알아볼 수 없도록 설계되었다.

## 1. 무유출(Zero-Leak) FDS 데이터 가공 및 전송 소스 코드

주피터 노트북 셀에 넣고 완성도를 높인 뒤, 최종 배포 시 파이참(`.py`) 파일로 변환하여 배치 파일에 연동하면 된다.

```python
import pandas as pd
import numpy as np
import uuid  # 출처 추적용 고유 난수 생성 라이브러리

# ==========================================
# [STEP 1] 관재인 내부 PC 데이터 로드 (가상 데이터)
# ==========================================
raw_data = {
    "거래일자": ["2026-03-05", "2026-03-12", "2026-04-01", "2026-04-15"],
    "고객명": ["홍길동", "홍길동", "홍길동", "홍길동"],
    "계좌번호": ["110-123-456789", "110-123-456789", "110-123-456789", "상가등기이전"],
    "적요내역": [" 마트 삼겹살 구매 ", " 자녀 홍이슬 증여 ", " 개인 채무 변제(사촌) ", " 소유권 이전 (배우자) "],
    "거래금액": ["45,000", "15,000,000", "8,000,000", "0"]
}
df = pd.DataFrame(raw_data)

# 내부 분석용 임시 전처리
df["정제적요"] = df["적요내역"].str.replace(" ", "")
df["정제금액"] = df["거래금액"].str.replace(r"[^\d]", "", regex=True).astype(int)


# ==========================================
# [STEP 2] 내부 분석 및 설명 가능한 룰 코멘트 부여
# ==========================================
# 관재인 PC 안에서만 돌아가는 룰 엔진 (외부 유출 안 됨)
conditions = [
    (df["정제금액"] >= 5000000) & (df["정제적요"].str.contains("자녀|아들|딸|증여")),
    (df["정제금액"] >= 5000000) & (df["정제적요"].str.contains("사촌|친척|지인|채무")),
    (df["정제적요"].str.contains("소유권|등기|이전") & df["정제적요"].str.contains("배우자|와이프"))
]

# 내부용 코멘트 및 머신러닝 학습용 코드(Rule ID) 동시 부여
comments = ["직계존비속 증여 의심", "친인척 편파변제 의심", "배우자 자산이전 포착"]
rule_ids = [101, 102, 201]  # 텍스트 대신 서버로 보낼 규칙 번호

df["분석코멘트"] = np.select(conditions, comments, default="정상")
df["Rule_ID"] = np.select(conditions, rule_ids, default=0)


# ==========================================
# [STEP 3] 제로-리스크 출처 추적용 해시 키(Trace_ID) 매칭
# ==========================================
# 이 프로그램이 실행될 때마다 관재인 PC 내부에서만 알 수 있는 일회성 고유 난수를 만든다.
# 본사 서버는 이 ID가 누구건지 절대 모르지만, 관재인 PC의 로그 파일과 대조하면 역추적이 가능하다.

trace_id = str(uuid.uuid4())[:8]  # 예: 'a1b2c3d4' 형태의 난수 생성
df["Trace_ID"] = trace_id


# ==========================================
# [STEP 4] 관재인용 데이터 vs 서버 전송용 데이터 분리
# ==========================================

# 1. 관재인 내부 PC에 저장되는 마스터 파일 (모든 정보 포함)
# 이 파일은 관재인 컴퓨터 하드디스크에만 저장되므로 안전하다.
관재인_보고서 = df[df["분석코멘트"] != "정상"][["거래일자", "고객명", "계좌번호", "적요내역", "거래금액", "분석코멘트", "Trace_ID"]]
관재인_보고서.to_excel("관재인_내부_마스터_파일.xlsx", index=False, engine="openpyxl")


# 2. [핵심] 본사 서버로 전송할 순수 통계 데이터 (텍스트 ZERO)
# 이름, 계좌, 적요 텍스트를 단 한 글자도 포함하지 않고 오직 숫자와 코드만 추출한다.
서버전송_데이터 = df[df["분석코멘트"] != "정상"][["Trace_ID", "거래일자", "정제금액", "Rule_ID"]]
서버전송_데이터.to_csv("pure_data_for_server.csv", index=False, encoding="utf-8-sig")

print("=== [성공] 온프레미스 제로-리스크 데이터 분리 완료 ===")
print("\n[서버로 전송되는 데이터 미리보기 (텍스트 전무)]")
print(서버전송_데이터)
```

### 📊 서버 전송 데이터 결과 화면 (`pure_data_for_server.csv`)
이 파일 내용이 나중에 중앙 머신러닝 분석 서버로 전송되는 순수한 통계 데이터이다.

# 온프레미스 배포를 위한 PyInstaller 단일 파일 실행 파일(EXE/APP) 패키징 가이드

파이썬 내장 패키징 도구인 **PyInstaller**를 사용하면, 내가 작성한 소스 코드(`.py`)와 Pandas 라이브러리 전체를 단 하나의 실행 파일(**윈도우용 `.exe` 또는 맥용 `.app`**)로 압축하여 패키징할 수 있다. 관재인은 파이썬 설치 없이 스마트폰 앱을 쓰듯 더블클릭만으로 FDS 프로그램을 즉시 가동할 수 있게 된다.

## 2. PyInstaller 설치 및 단일 파일 빌드 명령어

프로그램을 빌드할 때는 **실제 관재인이 사용할 운영체제와 동일한 환경**에서 진행해야 한다. (윈도우 PC에서 빌드하면 `.exe`가 나오고, 맥에서 빌드하면 맥용 실행 파일이 나온다.)

### 🛠️ 1단계: 패키징 라이브러리 설치
파이참 터미널 또는 명령 프롬프트(CMD) 창에서 아래 명령어를 실행하여 설치한다.

```bash
pip install pyinstaller
```

### 🛠️ 2단계: 단일 실행 파일 빌드 명령어 입력

# 온프레미스 단일 파일 배포를 위한 FDS 예외 처리(Try-Except) 보강 가이드

온프레미스로 배포된 실행 파일(`.exe`) 환경에서 가장 두려운 상황은 관재인이 무언가 잘못 조작했을 때 **프로그램이 아무런 설명 없이 팅겨서 꺼져버리는 현상(Crash)**이다. 주피터 노트북과 달리 파이참 기반 빌드 파일은 에러가 나면 검은색 콘솔 창이 0.1초 만에 닫히므로 어디가 고장 났는지 관재인도, 개발자도 알 수 없게 된다.

실무 현장에서 관재인이 저지를 수 있는 **3대 사용자 실수**를 정의하고, 프로그램이 강제 종료되지 않고 에러 메시지를 이쁘게 띄우며 대기하도록 만드는 단일 파일 방어 코드 구조이다.

---

## 1. 관재인 실무 현장의 3대 예외(Error) 상황
*   **Case 1 (파일 누락)**: 분석할 원본 엑셀 파일명을 잘못 적었거나 다른 폴더에 두고 실행한 경우 → `FileNotFoundError` 발생.
*   **Case 2 (컬럼명 오염)**: 은행에서 받은 엑셀 상단에 쓰레기 데이터가 있거나, 컬럼 이름이 '거래금액'이 아니라 '찾으신 금액' 등으로 되어 있는 경우 → `KeyError` 발생.
*   **Case 3 (타입 변환 실패)**: 금액 란에 숫자가 아닌 `"확인요망"`, `"공란"` 등의 텍스트가 적혀 있어 연산이 불가능한 경우 → `ValueError` 발생.

## 2. 결함 없는(Fault-Tolerant) FDS 엔진 통합 소스 코드

이 코드는 프로그램 전체를 하나의 커다란 `try-except` 블록으로 감싸고, 단계별로 예외 처리를 정교하게 분리하여 관재인에게 친절하게 가이드를 주도록 설계되었다.

```python
import pandas as pd
import numpy as np
import os
import sys

print("==================================================")
print("       보안 온프레미스 FDS 시스템 가동 (v1.0)       ")
print("==================================================\n")

# 프로그램 강제 종료를 막기 위한 전체 예외 처리 블록 시작
try:
    # ----------------------------------------------------
    # [예외 방어 1] 파일 존재 유무 및 파일명 정밀 체크
    # ----------------------------------------------------
    target_file = "파산자_은행내역.xlsx"

    if not os.path.exists(target_file):
        # 개발자가 직접 에러를 발생시켜 프로그램 흐름을 안전하게 하단 except로 유도
        raise FileNotFoundError(
            f"❌ 폴더 안에 '{target_file}' 파일이 존재하지 않는다.\n"
            f"   [조치 사항]: 분석할 엑셀 파일 이름을 정확히 수정하여 이 프로그램과 같은 폴더에 넣어라."
        )

    print(f"▶ [1단계] '{target_file}' 데이터를 로드하는 중...")
    # 엑셀 로드 시 발생할 수 있는 가벼운 엔진 에러 방어
    df = pd.read_excel(target_file, engine="openpyxl")


    # ----------------------------------------------------
    # [예외 방어 2] 필수 금융 데이터 컬럼 존재 여부 체크
    # ----------------------------------------------------
    required_columns = ["거래일자", "적요", "거래금액"]
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        raise KeyError(
            f"❌ 엑셀 파일 내부에 필수 컬럼{missing_columns}이 누락되었다.\n"
            f"   [현재 엑셀 컬럼]: {list(df.columns)}\n"
            f"   [조치 사항]: 엑셀 첫 번째 줄(헤더) 이름이 '거래일자', '적요', '거래금액'인지 확인하라."
        )

    print("▶ [2단계] 데이터 일괄 전처리 및 세척 진행 중...")
    # 데이터 전처리 실행
    df["정제적요"] = df["적요"].fillna("").astype(str).str.replace(" ", "").str.upper()


    # ----------------------------------------------------
    # [예외 방어 3] 금액 데이터 수치형 변환 실패 방어
    # ----------------------------------------------------
    # 숫자가 아닌 깨진 글자가 들어왔을 때 astype(int)를 쓰면 프로그램이 터진다.
    # 이를 방어하기 위해 정규식 정제 후 빈 값이 된 항목은 0원으로 안전하게 치환(fillna)한다.
    clean_amt_series = df["거래금액"].astype(str).str.replace(r"[^\d]", "", regex=True)

    # 공백이나 문자만 있어 숫자가 다 날아간 행은 '0'으로 채움
    clean_amt_series = clean_amt_series.replace("", "0")
    df["정제금액"] = clean_amt_series.astype(int)


    # ----------------------------------------------------
    # [3단계] 룰 기반 FDS 분석 및 필터링 수행
    # ----------------------------------------------------
    print("▶ [3단계] 관재인 설정 규칙 매칭 중...")
    target_amount = 3000000
    danger_keywords = "차남|와이프|지인|현금|가족"

    cond_amount = df["정제금액"] >= target_amount
    cond_keyword = df["정제적요"].str.contains(danger_keywords)

    df["FDS결과"] = np.where(cond_amount & cond_keyword, "🚨 고위험 은닉 의심", "🟢 정상")


    # ----------------------------------------------------
    # [4단계] 최종 보고서 출력
    # ----------------------------------------------------
    fds_report_df = df[df["FDS결과"] == "🚨 고위험 은닉 의심"]
    fds_report_df = fds_report_df[["거래일자", "적요", "정제금액", "FDS결과"]]

    output_excel = "FDS_관재인_최종보고서.xlsx"
    fds_report_df.to_excel(output_excel, index=False, engine="openpyxl")

    print("\n==================================================")
    print(f"🎉 [분석 성공] '{output_excel}' 파일이 안전하게 저장되었다.")
    print("==================================================")

# ----------------------------------------------------
# 시스템 에러 발생 시 관재인 화면 처리를 위한 예외 포착 구역
# ----------------------------------------------------
except FileNotFoundError as e:

    print("="*50)

# ----------------------------------------------------
```

## 💡 예외 처리 보강이 가져다주는 현업 효과

1. **프로그램 신뢰도 상승**: 관재인이 실수를 하더라도 DOS 창이 픽 꺼지는 것이 아니라, 하단 이미지처럼 `[파일 오류 발생] [조치 사항]: ...` 메시지가 명확하게 표기된 상태로 멈춰 서서 관재인을 기다려 준다. 관재인은 굳이 개발자를 부르지 않고도 스스로 엑셀 파일명을 고쳐서 문제를 해결할 수 있다.
2. **`fillna()`와 문자열 안전 변환** : 금액 데이터에 문자가 뒤섞여 들어오는 지저분한 환경에서도 프로그램이 중단되지 않고 악착같이 숫자를 정제하여 완주할 수 있는 탄탄한 맷집을 가진 엔진이 된다.
