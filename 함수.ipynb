#### 2. &nbsp; 일반적 함수의 예

```python
def sum_many(*args): # *args는 함수를 정의할 때 몇 개의 인자(Arguments)가 들어올지 모를 때 사용하는 문법
    sum = 0
    for i in args:
        sum += i # 혹은 sum = sum + i
    return sum
print(sum_many(1,2,3,))
```

```python
def sum_all(*args):
    # args가 내부적으로 어떤 모양인지 확인해보기
    print(f"들어온 값들의 형태: {args}")

    total = 0
    for num in args:
        total += num
    print(f"모든 값의 합은 {total}입니다.")

# 1. 3개의 인자 전달
sum_all(1, 2, 3)
# 출력:
# 들어온 값들의 형태: (1, 2, 3)  <- 튜플 형태로 묶임
# 모든 값의 합은 6입니다.

# 2. 5개의 인자 전달
sum_all(10, 20, 30, 40, 50)
# 출력:
# 들어온 값들의 형태: (10, 20, 30, 40, 50)
# 모든 값의 합은 150입니다.
```

```python
def print_kwargs(**kwargs):
    for k in kwargs.keys():
        if k == "name":  # k는 "name"(키)이고, kwargs[k]가 "int 조수"(값)이다.
           print("당신의 이름은 : " + kwargs[k])
print_kwargs(name="홍길동", c=3)
```

```python
def create_profile(user_id, **kwargs):
    print(f"=== [ID: {user_id}] 회원님의 프로필 ===")

    # 전달받은 추가 정보(kwargs)를 하나씩 꺼내어 출력
    for key, value in kwargs.items():
        print(f"• {key}: {value}")

    print("-----------------------------------")


# 상황 1: 추가 정보를 이름만 입력한 유저
create_profile("user01", name="홍길동")
# 상황 2: 추가 정보를 이메일과 사는 곳까지 입력한 유저
create_profile("user02", name="이영희", email="abc@email.com", location="서울")
```

#### 2-3 🔥함수 안에 선언된 변수의 효력 범위(중요)

```python
a = 1
def vartest(a): # 함수명부터 안의 출력까지의 범위 안에 있는 변수를 '지역변수'라고 한다. 지역변수는 지역에만 영향을 미친다.
    a = a+1     # 반면 전역변수는 프로그램 전체(모든 함수)에서 접근하여 사용 가능하다.
                # ← 이곳에 리턴(return)을 선언하지 않았다.
vartest(a)      # 결국 출력 값 1은 전역 변수 a=1에서 온 것이다.
print(a)
```

#### 2-3-1 global 키워드를 사용해서 return 없이 직접 수정하는 방법 알아보기

##### ➥ 함수 내부에서 global 키워드를 선언하면, 새로운 지역 변수를 만들지 않고 기존의 전역 변수를 직접 가리켜서 수정할 수 있다.

```python
a = 1  # 전역 변수

def vartest():
    global a  # 함수 바깥에 있는 전역 변수 a를 그대로 사용하겠다고 선언
    a = a + 1  # 전역 변수 a의 값을 직접 수정함

vartest()  # 함수 호출 (따로 인자를 넘기거나 return을 받지 않음)

print(a)  # 출력: 2
```

#### 📚&nbsp; 정리하기

- ##### global a는 "이 함수 안에서 a라는 이름을 쓰면, 그건 지역 변수가 아니라 바깥에 있는 전역 변수야"라고 파이썬에게 알려주는 역할을 한다.
- ##### 따라서 return을 사용해서 값을 전달하고 다시 대입하는 과정 없이도 전역 변수의 값이 직접 변경된다
- ##### 이때 &nbsp; 함수 정의 문에 &nbsp; ⚡️ def vartest(a): &nbsp; 처럼 인자를 넣으면 안 된다. 매개변수 a는 지역 변수로 취급되므로 global a와 충돌이 일어나기 때문이다.

#### 🏢&nbsp; 실무에서는 어떻게 사용할까?
##### 1. 리턴(return) 활용 방식값을 함수 밖으로 전달하여 전역 변수를 업데이트하는 방식이다.
- ##### 장점 (독립성과 재사용성): 함수가 외부 변수에 의존하지 않아 언제든 다른 프로그램에서 재사용할 수 있다.
- ##### 장점 (예측 가능함): 입력값에 따른 출력값이 명확하므로 버그를 찾기 쉽고 테스트 코드를 짜기 좋다.
- ##### 단점: 결과값을 매번 변수에 다시 대입(a = vartest(a))해 주어야 하므로 코드가 약간 길어질 수 있다.

##### 2. global 키워드 방식: 함수 내부에서 전역 변수를 직접 제어하는 방식이다.
- ##### 장점 (간결함): return과 대입 과정이 필요 없어 코드가 일시적으로 짧아 보인다.
- ##### 단점 (스파게티 코드 위험): 여러 함수가 동시에 global a를 수정하면, 어떤 함수 때문에 a 값이 바뀌었는지 추적하기 매우 힘들어진다.
- ##### 단점 (재사용 불가): 해당 전역 변수가 바깥에 선언되어 있지 않으면 이 함수는 작동하지 않으므로 독립성이 깨진다.

##### 💡 실무 결론: 코드가 조금 더 길어지더라도 안전하고 깔끔한 return 방식을 사용하자.
##### 💡 global은 프로그램의 규모가 커질수록 치명적인 버그의 원인이 된다.

#### 2-4 𝜆 람다(Lambda)함수

```python
def add(a,b):
    return a+b
# add = lambda a,b: a+b - 이게 람다 함수다.
print(add(1,2))
```

##### 𝜆 &nbsp; 람다 함수(익명함수)

##### 𝜆 람다 함수(익명함수) ➝ lambda 매개변수 : 표현식


###### 특징:
###### def 키워드가 사라진다.
###### return 키워드를 쓰지 않아도 표현식의 결과가 자동으로 반된다

```python
add = lambda a,b: a+b
print(add(1,2))
```

##### 심지어 𝜆 람다 함수(익명함수)는 List &nbsp;안에서도 사용할 수 있다.

```python
myList = [lambda a,b: a+b, lambda a,b: a*b]
print(myList[0](1,2))
print(myList[1](2,2))
```
