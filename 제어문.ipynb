#### 2 &nbsp; 반복문(while/ for)

##### 2-1 while <조건문>: &nbsp; Bool 자료형이 나올 수 있도록 만들어주는 곳(True or False)
##### <수행할 문장>
##### <수행할 문장2>
##### <수행할 문장3>
....

```python
treeHit = 0
while treeHit <10: # while 조건이 True라면 아래 조건을 계속해서 반복해서 싫행한다.
    treeHit = treeHit + 1
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어 갑니다.") # if 문에 따른 결과를 만들기 때문에 print문은 들여쓰기를 해줭야 한다.
```

##### 2-2 &nbsp;  f-string으로 더 간결하게 print 함수를 만들 수 있다.

```python
treeHit = 0
while treeHit < 10:  # while 조건이 True라면 아래 조건을 계속해서 반복해서 싫행한다.
    treeHit = treeHit + 1 # 이 부분도 treeHit += 1 이렇게 간략하게 만들 수 있다.
    print(f"나무를 {treeHit}번 찍었습니다.")
    if treeHit == 10:
        print("나무 넘어 갑니다.")
```

##### 🔥Quiz 1!:
##### &nbsp;&nbsp;&nbsp;아래 예제는 본질적으로 &nbsp; *'복리 계산법'* 에 해당하는 간단한 *'지수적 성장 모델'* 을 이용해 박테리아 집단의 성장률을 계산하는 문제다.
##### &nbsp;&nbsp;&nbsp;&nbsp;*P(t +1) = P(t) +rP(t)* 이 공식에서 &nbsp; *P(t)는&nbsp; t* 시간의 개체 수 규모이고, &nbsp;*r은* 성장률이다. 이제 박테리아 집단이 처음의 두 배가 되는
##### &nbsp;&nbsp;&nbsp;&nbsp;데 걸리는 시간을 알아보자.

```python
time = 0
population = 1000 #1000개의 박테이라로 시작
growth_rate = 0.21 #21%의 분당 성장률
while population < 2000:
    population = population + growth_rate * population
    print (population)
    time = time + 1
print(f"박테리아 수가 두 배가 되는 데 걸린 시간은 {time}분 입니다.")
print(f"... 그리고 최종 박테리아 수는 {population}개 입니다.")
```

##### 🔥Quiz 2: 무한 푸프 (inf)
##### 앞의 예제는 Population < 2000을 조건으로 썼으므로 루프는 박테리아 수가 처음에 비해 두 배 이상에 도달하면 멈춘다. 그럼 박테리아 수 가 정확히 처음에 비해 두 배일 때만 멈춘다면 어떻게 될까?

```python
time, population,growth_rate = 0, 100, 0.21 # 다중값 할당을 이용한 제어용 변수 초기화한다.
# threshold, max_iter, learning_rate = 0.5, 100, 0.01 이처럼 FDS 작업할 때는 모델 파라미터 초기값 설정할 때 이런 식으로 자주 쓴다.
while population != 2000:
    population = population + growth_rate * population
    time = time + 1
    if time >20:
        print(f"정확히 2000이 되는 경우는 없습니다. {time}번 반복 후 중단.")
        break
# print문 아래에 break가 없기 때문에 무한 반복(inf)을 멈추지 못한다. 강제로 멈추기 위해 command +c나 실행버튼을 stop버튼을 눌러 멈추게 했다.
# 결과 값은 정확히 2000이 되는 지점을 만들지 못하므로 무한 루프에 빠지게 된다. 즉 모든 결과 값에 2000이 되는 수는 없다.
```

```python
marks = [90,25,67,45,80]
number = 0
for mark in marks:
    number = number + 1
    if mark >= 60:
        print(f"{number}번 학생은 합격 입니다.")
    else:
        print(f"{number}번 학생은 불합격 입니다.")
```

#### 3-2 &nbsp; for 문과 continue

```python
marks = [90,25,67,45,80]

number= 0
for mark in marks:
    number = number + 1
    if mark < 60:
        continue # continue에 걸리면(True) 처음으로 돌아가고, 안 걸리면(False) 통과하여 출력한다.
    print(f"{number}번 학생 축하합니다. 합격입니다.")
```

#### 3-5 ★ 리스트 내포(List comprehension)

- ##### 리스트 내포(List Comprehension)는 파이썬(Python) 등에서 기존 리스트를 기반으로 새로운 리스트를 간결하고 직관적으로 생성하는 기법 이다.
- ##### 여러 줄의 for 반복문이나 if 조건문을 대괄호(\([]\)) 한 줄로 압축하여 작성할 수 있어 가독성이 좋고 실행 속도가 빠르다.

#### 기본구조
##### 가장 기본적인 형태는 반복할 요소를 표현식으로 처리하는 것이다.

##### 1. &nbsp;&nbsp;표현식 for 항목 in 반복 가능한 객체

##### 예시: 0부터 9까지의 숫자를 각각 제곱한 리스트 생성

```python
squares = [x**2 for x in range(10)]
print(squares)
```

```python
result = [num *3 for num in a] # 앞의 num은 뒤에 나오는 for 문을 돌린 결과다.

result = []
for num in a:
    result.append(num * 3)

print(result)
```

##### 2. &nbsp;&nbsp;조건문 포함 (필터링)for문 뒤에 if문을 추가하여 특정 조건에 맞는 요소만 추출할 수 있다

##### 3. &nbsp;&nbsp; 표현식 for 항목 in 반복 가능한 객체 if 조건문


```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
# 결과: [0, 4, 16, 36, 64]
print(even_squares)
```

```python
a = [1, 2, 3, 4, 5]
result = [num *3 for num in a if num % 2 == 0]

result = []
for num in a:
    if num % 2 == 0:
        result.append(num * 3)
print(result)
```

##### 4. &nbsp;&nbsp; 조건-분기 포함 (if/else)

##### &nbsp; &nbsp; if / else를 모두 사용할 경우, 조건문을 for문 앞으로 이동시켜 작성한다

##### &nbsp;&nbsp;&nbsp;&nbsp;[참일_때_표현식 if 조건문 else 거짓_때_표현식 for 항목 in 반복_가능한_객체]

- ##### 예시: 짝수는 제곱하고, 홀수는 그대로 두기

```python
mixed = [x**2 if x % 2 == 0 else x for x in range(10)]
# 결과: [0, 1, 4, 3, 16, 5, 36, 7, 64, 9]
print(mixed)
```
