#### 3. 아래 문제를 보고 변수의 구조를 잘 이해하도록 하자. (잠시 곰곰히 생각해보라.)

##### 3-1. 얼핏 보면 변수 a 의경우 변화된 값을 갖는게 자연스럽지만 b의 경우 영향을 받지 않고 그대로 유지하고 있는게 맞지 않나?
##### 라고 생각을 할 수 있다. 하지만 그렇지 않다. a도 b도 모두 같은 주소를 바라보고 있는 것이다. b = a 라고 했을 때 이것의 의미는
##### b에 a의 주소를 넣는다는 의미다.
##### 즉, 주소를 넣은 것이지 값을 넣은 건 아니다. 변수(주소)를 가르키는 영역과 값을 가르키는 영역을 헷갈리지 말아야 한다.

```python
a  = [1,2,3]
b = a
a[1] = 4
print(a)
print(b)
```

 ##### 3-2 &nbsp; 다음 예제를 보면 위 예제와 다른점을 볼 수 있다.
 ##### &nbsp;&nbsp; &nbsp;&nbsp; &nbsp;&nbsp;&nbsp;a와 b의 주소값이 같다. &nbsp; id()함수는 주소를 찾는 함수다. &nbsp;&nbsp; Print (a is b)도 주소가 같은지를 묻는 프린트 함수다.

```python
a = [1,2,3]
b =a
a[1] = 4
print(id(a))
print(id(b))
print(a is b)
```

##### 3-3 &nbsp;&nbsp; b에 a의 값(value)을 주고 싶을 때는 어떻게 해야 한느가? ➞ 슬라이싱을 사용한다. 즉 우변에는 변수와 값을 포함하는 것이 올 때 &nbsp; b는 a의 값까지도 받게 된다.
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 그럼 이 경우 b의 주소와 값이 들어 있는 메모리도 달라진 것인가?
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 맞다. b=a는 같은 상자에 이름표만 두 개 붙인것이고 b=a[:]는 b라는 상자에 [1,2,3]을 넣은 것이다. 주소가 달라졌다.
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 반면 a[1]=4의 의미는 a라는 방에 2를 빼고 4를 넣은 것이다. 즉 주소는 여전히 같은데 값만 변한 것이다.

```python
a = [1,2,3]
b = a[:]
a[1] = 4
print(b)
print(a)
print(id(a))
print(id(b))
```

## 🔥 3-4. 중첩 리스트와 복사의 함정 (얕은 복사: Shallow Copy)

### 1) 질문: `a = [1, [2, 3], 4]` 일 때, `b = a[:]`로 슬라이싱 복사를 한 후 `a[1][0] = 99`를 하면 `b`는 어떻게 될까?
* **예측:** `b`는 독립된 상자이므로 값의 변화 없이 원래 상태 `[1, [2, 3], 4]`를 유지할 것이다?
* **결과:** **틀렸다.** `b` 역시 `[1, [99, 3], 4]`로 함께 변경된다.

---

### 2) 왜 이런 현상이 일어나는가? (파이썬의 메모리 관리 원칙)
리스트 상자는 두 개 이상의 메모리 주소를 한 칸에 직접 담지 못한다. 따라서 파이썬은 중첩된 리스트를 처리할 때 다음과 같은 방식을 사용한다.

* **실제 메모리 구조 (주소 쪽지 개념):**
  * `a` 상자의 첫 번째 칸: 숫자 `1`
  * `a` 상자의 두 번째 칸: 독립된 다른 방에 있는 `[2, 3]`의 *'주소 쪽지(참조)'* 가 들어있음.
  * `a` 상자의 세 번째 칸: 숫자 `4`

* **슬라이싱(`b = a[:]`)을 했을 때의 내부 동작:**
  * 파이썬은 겉 표면의 새 상자 `b`를 만들고 `a` 내부의 내용물들을 복사한다.
  * 이때, 두 번째 칸에 있던 **'주소 쪽지'가 그대로 `b`로 복사**된다.
  * 결과적으로 `a`와 `b`는 서로 다른 겉 상자를 갖지만, 내부의 두 번째 칸은 **동일한 주소 쪽지**를 들고 같은 `[2, 3]` 방을 바라보게 된다.

---

### 3) 메모리 시각화 흐름

1. **초기 상태 및 슬라이싱 복사 (`b = a[:]`)**
   * `a` ──> [ 1 | 주소 쪽지(0x101) | 4 ]
   * `b` ──> [ 1 | 주소 쪽지(0x101) | 4 ]
                  │
                  └───> [2, 3] (메모리 주소: 0x101)

2. **값 변경 발생 (`a[1][0] = 99`)**
   * `a`가 가진 주소 쪽지(`0x101`)를 타고 들어가 첫 번째 값을 `99`로 수정한다.
   * 공유 방의 데이터가 바뀌었으므로, `b`가 주소 쪽지를 타고 들어가도 똑같이 바뀐 값을 보게 된다.
   * `a` ──> [ 1 | 주소 쪽지(0x101) | 4 ]
   * `b` ──> [ 1 | 주소 쪽지(0x101) | 4 ]
                  │
                  └───> [99, 3] (메모리 주소: 0x101)

---

### 📌 핵심 요약 (FDS 프로젝트/데이터 분석 주의사항)
* 리스트 슬라이싱(`[:]`)이나 `.copy()`는 가장 바깥쪽 상자만 새로 만드는 *'얕은 복사(Shallow Copy)'* 이다.
* 내부에 또 다른 가변 객체(리스트, 딕셔너리 등)가 존재한다면 내부 주소까지 복사하지 못하고 공유하게 된다.
* 데이터 분석 시 원본 데이터를 보존하면서 다차원 배열을 안전하게 복사하려면 내부 알맹이까지 모두 새로 만드는 *'깊은 복사(Deep Copy)'* 가 필요하다.

```python
a = [1,[2,3],4]
b = a[:]
a[1][0] = 99
print(a)
print(b)
print(id(a))
print(id(b))
print(id(a[1]))
print(id(b[1]))
```

* **최종 검증 (id 확인):**
  * `id(a) == id(b)` ➞ `False` (겉 상자의 주소는 서로 다르다!)
  * `id(a[1]) == id(b[1])` ➞ `True` (내부 알맹이 리스트의 주소는 같다!)

  "겉 상자는 슬라이싱으로 새로 만들어졌지만, 내부 리스트는 주소 쪽지만 복사되어 같은 곳을 바라보고 있기 때문에 발생하는 현상이다."

```python
a = [1,[2,3],4]
b = a[:]
a[1][0] = 99

id(a) == id(b)
```

```python
a = [1,[2,3],4]
b = a[:]
a[1][0] = 99

id(a[1]) == id(b[1])
```

```python
from copy import copy # copy()라는 함수를 가져왔기 때문에 아래와 같은 결과가 나온다.
a = [1,2,3]
b =copy(a)
a[1] = 4
print(a)
print(b)
print(id(a))
print(id(b))
```

### 🔥값의 교환(swapping) 다른 언어와 달리 파이썬은 매우 직관적으로 스와핑 한다.

```python
a = 3
b = 5
a,b= b,a
print(a)
print(b)
```

#### 🔥다른 언어의 경우 임시 저장소인 temp 라는 걸 거쳐야 스와핑(swapping)이 가능하다.

```python
a = 3
b = 5
temp = b
b = a
a = temp
print(a)
print(b)
```
