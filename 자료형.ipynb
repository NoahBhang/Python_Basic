##### 슬라이싱의 경우 [ : ] 의 : 뒤에 오는 숫자 앞까지 잘라준다는 의미임을 잊지 말자. 즉, a[ : : : ]  이 있을 때
##### 첫번째 :공백은 *~이상* 두번째: : 공백은 *~미만* 으로 이해하면 된다. 그리고 : : : 마지막은 *간격*을 의미한다.
##### 0<= x <4

```python
a = '20010331Rany'
print(a[ :8]) # 처음부터 8미만까지 출력
print(a[8:]) # 8부터 마지막까지 출력
print(a[0: : 2]) # 두 칸씩 이동하여 출력한다는 의미. 처음부터 마지막까지 두칸 씩 이동하여 출력
```

#### 해당 코드는 파이썬의 문자열 포매팅(String Formatting) 중 % 연산자를 사용하여 문자열 안에 원하는 값을 동적으로 삽입하는#### 방법을 설명하는 예제.
#### 각 부분의 상세 역할은 다음과 같다.
- ##### %d: 정수(Integer) 가 들어갈 자리를 지정하는 포맷 코드이다.
- ##### % 3: 문자열 뒤에 있는 % 연산자가 앞의 문자열을 탐색하며, %d 위치에 숫자 3을 대입합니다.
- ##### print(a): 변수 a에 저장된 최종 문자열 'I eat 3 apples.'를 화면에 출력합니다. 응근이 많이 쓰이는 유용한 문법이다.

```python
a = 'I eat %d apples.' % 3
print(a)
b = 'I eat '+ str(3) + ' apples.'
print(b)
```

- #### %d 는 정수를 받을 경우 사용하고 %s는 문자열을 받을 때 사용한다. %d를 유지하고 싶다면 'three'가 아니라, 3을 써야한다.

```python
number = 10
day = 'three'
a = 'I ate %d apples. so I was sick for %s days.' % (number, day)
print(a)
```

### 🔥최근 파이썬(3.6 버전 이상)에서는 % 연산자 대신 변수 이름을
*중괄호 {} 안에 변수명을 직관적으로 넣을 수 있는 f-문자열(f-string)*  방식을 가장 많이 사용한다.
이 방식을 쓰면 타입 고민 없이 훨씬 안전하고 편리하게 코드를 작성할 수 있다. 따라서 %d,%s,%f 등을 이제 사용할 필요가 없다. 하지만 레거시 코드분석이나 c언어의 경우 여전히 사용하므로 %d는 decimal(정수), %s는 string, %f는 floating의 약자임을 기억해 놓자.

```python
number = 10
day = 'three'
# 문자열 앞에 f를 붙이고 변수를 {}에 삽입
a = f'I ate {number} apples. so I was sick for {day} days.'
print(a)
```

##### 아래 슬라이싱에서 마지막 간격을 의미하는 *[-1]* 의 의미는 맨 뒤에서부터 1간격씩 출력하라는 의미다.

### *if문* 이랑 결합해서 글자가 있으면 어떤걸 처리하고, 없으면 어떤걸 처리하고. 이런 경우가 많기 때문에 *.find()* 는 자주 사용된다.

```python
a = 'Python is best choice'
a.find('b')#알파벳 b를 어디에서 찾을 수 있냐? [10]에 위치해 있다고 알려줌.
```

```python
a = 'Python is best choice'
a.find('k') # 없는 걸 찾으면 -1이 나온다.
```
