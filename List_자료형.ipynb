### 바다 코끼리 연산자, 왈러스 연산자라고 부른다.
#####  *:= 는   왈러스 연산자(walrus operator)* 라고 부르고, "값을 변수에 대입하면서 동시에 그 값을 반환"하는 연산자이다.
#### 핵심은 오른쪽에 반드시 값(표현식)이 와야 한다는 점이다. a := 처럼 오른쪽이 비어 있으면 오류가 난다.

```python
print(a := [1, 2, 3])   # a에 [1,2,3]을 대입 + 출력 → [1, 2, 3]
print(b := a[0])        # b에 a[0](=1)을 대입 + 출력 → 1
```

#### *바다 코끼리  연산자(walrus operator)* 는 실무에서 주로 if나 while, 컴프리헨션 안에서 *"값을 계산해서 변수에 담는 동시에 조건 검사"* 를 할 때 유용하다.
#### 예를 들어:

```python
# 한 줄에서 계산 + 검사
if (n := len(a)) > 2:
    print(f"리스트 길이가 {n}개라 충분합니다")
```

#### 여러 개의 값을 제거하기 위해서는 *'for문'* &nbsp; 등을 사용해야 한다.
##### 1. 리스트 컴프리헨션 사용 *☆☆(가장 추천)조건문(if)☆☆을 활용* 해 삭제할 값을 제외한 나머지 데이터만 모아서 새 리스트를 만드는
##### 방법이다. 코드가 깔끔하고 성능이 좋다.

```python
a = [1, 2, 3, 1, 2, 3]

# 3이 아닌 값들만 남기기
a = [x for x in a if x != 3]

print(a)  # 출력: [1, 2, 1, 2]
```

##### 2. filter() 함수와 lambda 사용내장 함수인 filter와 if 조건 역할을 하는 lambda를 조합하여 특정 값을 걸러내는 방법이다.


```python
a = [1, 2, 3, 1, 2, 3]

# 3이 아닌 요소만 필터링
a = list(filter(lambda x: x != 3, a))

print(a)  # 출력: [1, 2, 1, 2]
```

#### 동일한 내용을 ☆ *'바다 코끼리 연산자'* ☆&nbsp;를 사용해 만들어 보자.
#### 이 경우 바다 코끼리 연산자가 *'버그(bug)'를 줄이는 데* 더 유리하다.
##### 전자 .extends는 기존에 [1, 2, 3]이 들어있던 메모리 방(주소)을 그대로 유지한 채, 그 방의 크기를 늘려서 4, 5를 이어 붙인다.
##### 후자 *(바다 코끼리 연산자(:=))* 의 경우 컴퓨터가 메모리의 완전히 새로운 기슭에 [1, 2, 3, 4, 5]라는 새로운 방을 만듭니다. 그리고 변수 이름 a가 원래 방을 버리고 이 새로운 방을 가리키도록 이사(재할당)를 보낸다.

```python
a = [1, 2, 3]

# 변수 a에 [4, 5]를 더한 값을 대입하면서, 동시에 print로 출력한다.
print(a := a + [4, 5])

# 출력 결과: [1, 2, 3, 4, 5]
```
